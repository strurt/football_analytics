```python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('PastDecade.csv')
```

```python
data
```

```python
data['GD'] = data['GF'] - data['GA']
```

```python
data
```

```python
x=data['GD']
y=data['Pts']
```

```python

linreg=LinearRegression()
x=x.values.reshape(-1,1)
linreg.fit(x,y)
y_predict=linreg.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict,color='red')
plt.title=('Goal difference vs points')
plt.xlabel('GD')
plt.ylabel('Points')
```

```python
slope=linreg.coef_
intercept=linreg.intercept_
Rsqr=linreg.score(x,y)
```

```python
print("Slope=",slope,"intercept=",intercept,"R squared=",Rsqr)
```
